```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

```python
df = pd.read_csv(r'sample_superstore.csv', encoding='windows-1252')
```

## Análise Inicial

```python
display(df.head(5))
```

```python
df.info()
```

```python
display(df.describe())
```

```python
# verificando valores nulos
missing_values = df.isnull().sum()
display(missing_values[missing_values > 0])
```

```python
# transformando colunas para nomes minúsculos e substituindo espaços
df.columns = df.columns.str.lower().str.replace(' ','_').str.replace('-','')
display(df.columns)
```

## Criando a tabela customer

```python
customer = df.loc[:,['customer_id','customer_name','segment']].drop_duplicates()
display(customer)
```

## Criando a tabela sales

```python
sales = df.loc[:,['order_id','order_date', 'ship_date', 'ship_mode','customer_id','postal_code','product_id']]
display(sales)
```

## Criando a tabela location

```python
location = df.loc[:,['country', 'city', 'state','postal_code', 'region']].drop_duplicates()
display(location)
```

## Criando a tabela product

```python
product = df.loc[:,['product_id', 'category', 'subcategory','product_name']].drop_duplicates()
display(product)
```
